```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("ECOM.csv")
```

```python
data.head() #first 5 row and column to visualize how the data looks like
```

```python
x = data.iloc[:,3:7].values #independent 
```

```python
x
```

```python
y = data.iloc[:,7:].values #dependent
```

```python
y
```

```python
X = data.iloc[:,6:7].values
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(X,y,color='blue')
```

```python
from sklearn.model_selection import train_test_split #importing train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) #splitting data into train and test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
y_predict[3]
```

```python
y_test[3]
```

```python
#so the model is pretty accurate as predicted,but lets see
```

```python
print("Accuracy=", (r2_score(y_test, y_predict)*100))
```
